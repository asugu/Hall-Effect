```python
import numpy as np
from ROOT import TCanvas, TGraphErrors, TF1, TGraph,TLatex
from ROOT import gROOT
from array import array
```

```python
Ve=[6.59,5.71,3.19,6.98,4.39,2.35,4.22,6.0,7.45,8.45]
Ae=[125.1,107,62.2,132,85.7,46.4,84.5,118,141.5,160]
V=[-14.30,-13.40,-12.00,-10.40,-11.60,-12.80,-14.80,-15.70,-16.70,-15.10]
B=[147.50,134.80,109.70,82.80,104.80,127.40,162.40,175.30,192.20,163.60,]
Verr=[]
Berr=[]
ndata=len(V)

for i in range(0,ndata):
    Verr.append(0.1)
    Berr.append(B[i]*0.05)



```

```python
V  =  array( 'f', V)
A  =  array( 'f', B)
Verr= array('f',Verr)
Aerr=array('f',Berr)
```

```python
c1 = TCanvas( 'c1', '..', 200, 10, 700, 500 )
             
mygraph = TGraphErrors(ndata,A,V,Aerr,Verr)
mygraph.Draw("AP")
mygraph.SetTitle("Reverse Hall voltage versus Magnetic field fixed at 117.60mA")
mygraph.SetLineColor( 1 )
mygraph.SetLineWidth( 2 )
mygraph.SetMarkerColor( 1 )
mygraph.SetMarkerStyle( 4 )

mygraph.GetXaxis().SetTitle( 'Magnetic field (mT)' )
mygraph.GetYaxis().SetTitle( 'Voltage (mV)')
c1.Draw()
```

```python
l1=TF1('l1','pol1')
mygraph.Fit(l1,'','')
c1.Draw()
```
